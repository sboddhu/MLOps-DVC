# DVC Collaboration Tutorial (Colab)

This notebook runs the **same pipeline** as the repo and shows a fast collaboration workflow.

**In-class idea:** students fork the repo, run baseline, then run experiments via `dvc exp run`.


````
# (1) Clone your repo
# Replace with your fork URL, or upload the zip to Colab and unzip.
REPO_URL = "https://github.com/<your-org-or-user>/<your-repo>.git"
!git clone {REPO_URL} repo
%cd repo
````

````
# (2) Install deps
!pip -q install -r requirements.txt
````

````
# (3) Initialize DVC (safe if already initialized)
!dvc init -f
````

````
# (4) Run the pipeline
!dvc repro
````

````
# (5) View metrics
!dvc metrics show
````

````
# (6) Quick experiments (no commits required)
!dvc exp run -S train.C=0.2
!dvc exp run -S train.C=2.0
!dvc exp show
````

````
# (7) Compare experiments
!dvc exp diff
````
